```python
import networkx as nx
import nxv

# Create a directed networkx graph
G = nx.DiGraph()

G.add_node("ackrep_core")
G.add_node("ackrep_data", shape="rect", color="#909090")
G.add_node("ackrep_data\n(for unittests)", shape="rect", color="#909090")
G.add_node("pyerk_core")
G.add_node("pyerk_django")
G.add_node("pyerkconf.toml", shape="rect")
G.add_node("execscript.py", shape="rect", color="red")
G.add_node("ocse", shape="rect", color="blue")
G.add_node("ocse\nfor unittests", shape="rect", color="blue")
G.add_node("config.ini", shape="rect", color="green")

G.add_edge("ackrep_core", "ackrep_data")
G.add_edge("ackrep_core", "ackrep_data\n(for unittests)")

G.add_edge("ackrep_core", "pyerk_core")
G.add_edge("pyerk_core", "pyerk_django")
G.add_edge("pyerk_core", "pyerkconf.toml")
G.add_edge("pyerk_django", "pyerkconf.toml")
G.add_edge("ackrep_data", "execscript.py", style="dashed")
G.add_edge("execscript.py", "ackrep_core")
G.add_edge("ackrep_core", "ocse")
G.add_edge("ackrep_data", "ocse")
G.add_edge("ocse", "pyerkconf.toml", label="provides")
G.add_edge("ackrep_core", "config.ini")
G.add_edge("config.ini", "ocse", label="path to")
G.add_edge("ackrep_data\n(for unittests)", "ocse\nfor unittests")


style = nxv.Style(
   graph={"rankdir": "BT"},
   node=lambda u, d: {"shape": d.pop("shape", "ellipse"), **d},
   edge=lambda u, v, d: {**d},
)

nxv.render(G, style)
```
